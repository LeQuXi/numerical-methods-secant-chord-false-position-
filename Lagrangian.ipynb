```python
import numpy as np
import matplotlib.pyplot as plt
def func(x):
  return np.cos(2 * ùë•)* (np.sin(2 * ùë•) + 1.5)-(x/5)
n = 10
Min= -2
Max = 3

unix = (Max-Min)/(n-1)
X= np.arange(Min, Max+unix, unix)
Y = func(X)
plt.plot(X,Y, 'ro')
n= len(X)
def Lagrangian(x, X, j):
  n = len(X)
  L = 1
  for i in range(n):
    if i == j: 
      continue
    else:
      L = L*(x-X[i])/(X[j]-X[i])  
  return L
xxx = np.arange(Min, Max, 0.1)
yyy = np.zeros_like(xxx)
print(yyy)
for i in range(n):
    yyy_l = Lagrangian(xxx, X, i)
    #plt.plot(xxx, yyy_l)
    yyy += Y[i]*yyy_l

print(yyy_l)
plt.grid()
plt.plot(xxx, yyy, 'b-', linewidth = 3)
plt.plot(xxx, func(xxx), 'g-', linewidth = 2)
plt.plot(xxx, yyy-func(xxx), 'r-')
plt.show()
```
